Create a class called FootballTeam. Each football team will have a few variables:

teamName is a string representing the name of the team.

team is a list of players denoted by their name in string.

strength is an float number which corresponds to how strong a team is upon a 100 scale.

manager is the name of the team manager in string. 

record is a list of string which represents their winning record. It should only contain either of 3 values: 'W','L','D'

Each FootballTeam class has 4 methods:

form() generates a random float value between 0 to 100, and updates it as the strength variable. If there is no manager in the team, form() generates a float value between 0 to 70. If the last known value in record is 'W', the strength is increased by 10%. If the last known value in record is 'D' or there is no prior record, there is no change in the strength. If the last known value in the record is 'L', the strength is decreased by 10%.

updateRecord() updates the record list of the team with the latest result. It takes in the result of the match in its parameter, and returns nothing.

fire() removes the manager and replaces it with None if the last 3 values in the team's record is 'L'. 

hire() checks if there is no manager in the team, and generates a random number between 1 to 3. If the value is 2 and there is no manager in the team, the team hires a manager called "Jorden".

Create a class called FootballLeague. Each football league enrolls 5 FootballTeams in it's constructor, and stores it in a variable called teams.

FootballLeague has a dictionary called results. Results stores a key-value pairing of each team and their score. Each win gives 3 points, a draw gives 1 point and a loss gives 0 points.

FootballLeague has a 2 methods:

play() takes in two teams and calls the form() method of each team, before comparing the strength of each team. The team with the greater strength wins. play() also updates the results dictionary with the corresponding scores. 
play also calls the updateRecord() method of each football team, and checks if there is a need to fire or hire a manager.

season() plays each team with all other teams using the play() method. It takes in the the year of the season in integer, and returns a Winner object.

Winners have 2 methods:

newWinner()takes in the name of the team that won, and the year of the season. It then adds it into the winners list.

currentWinner() returns the name of the latest team to have won the league.

```python
import random
class FootballTeam: 
    def __init__(self,teamName,team,manager):
        self.teamName = teamName
        self.team = team
        self.strength = 0
        self.manager = manager
        self.record = []
    
    def form(self):
        self.strength = random.uniform(0,100)
        if len(self.manager) == 0 or self.manager == "None": 
            self.strength = random.uniform(0,70)
        elif len(self.record) == 0 or self.record[-1] == "D": 
            self.strength = self.strength
        elif self.record[-1] == "W":
            self.strength = self.strength * 1.1
        elif self.record[-1] == "L":
            self.strength = self.strength * 0.9
        return self.strength
    
    def updateRecord(self,latestResult): 
        self.record.append(latestResult)
        
    def fire(self): 
        if self.record[-1] == "L" and self.record[-2] == "L" and self.record[-3] == "L": 
            self.manager = "None"
        return self.manager
            
    def hire(self): 
        randomNumber = random.randint(1,3)
        if randomNumber == 2 and self.manager == "None":
            self.manager = "Jorden"
        return self.manager
    
class FootballLeague: 
    def __init__(self,team1,team2,team3,team4,team5):
        self.teams = [team1,team2,team3,team4,team5]
        self.results = {team1.teamName:0,team2.teamName:0,team3.teamName:0,team4.teamName:0,team5.teamName:0}
     
    def updateScore(self,fc,scoreEarned): 
#Retrieve current score of the team       
        currentScore = self.results[fc.teamName]  
#Update new score of team
        newScore = currentScore + scoreEarned
#Assign new score of team
        self.results[fc.teamName] = newScore    
            
    def play(self,team1,team2):
        team1Strength = team1.form()
        team2Strength = team2.form()
        winning = "W"
        losing = "L"
        drawing = "D"
        if team1Strength > team2Strength: 
            team1.updateRecord(winning)
            self.updateScore(team1,3)
            team2.updateRecord(losing)
            self.updateScore(team2,0)
        elif team1Strength < team2Strength: 
            team1.updateRecord(losing)
            self.updateScore(team1,0)
            team2.updateRecord(winning)
            self.updateScore(team2,3)
        else: 
            team1.updateRecord(drawing)
            self.updateScore(team1,1)
            team2.updateRecord(drawing)
            self.updateScore(team2,1)
        team1.hire()
        team2.hire()
        
    def season(self,yearOfSeason):
        for i in range(len(self.teams)):
            self.results[self.teams[i].teamName] = 0
#         print("debugging results") 
#         print(self.results)
        for team1 in self.teams: 
            for team2 in self.teams: 
                if team1 != team2:
                    self.play(team1,team2)
        import operator
        winner = max(self.results.items(), key=operator.itemgetter(1))[0]
        return winner

class Winner(FootballTeam): 
    def __init__(self,championships):
        self.championships = championships
        if winner == FootballTeam:
            self.championships.append(yearOfSeason)
    def swag(self):
        print("I, " + str(self.teamName) +" have won " + len(self.championships) + "championships.")
        print("I won the championship on" + self.championships + ".")
        if FootballTeam == winner:
            print("I am the current winner.")

class Winners:
    def __init__(self):
        self.winners = []
    def newWinner(self,nameOfTheTeam,yearOftheSeason):
        self.winners.append((yearOftheSeason,nameOfTheTeam))
    def currentWinner(self):
        return winner
```

```python
# Explaining classes and objects
class Cookie:
    def __init__(self):
        self.percentLeft = 100
        
    def bite(self):
        if self.percentLeft == 0:
            print("There is nothing left to eat of this cookie")
        else:
            self.percentLeft -= 10
            print("There is " + str(self.percentLeft) + "% of the cookie left to eat")
            
    def eat(self):
        self.bite()
```
